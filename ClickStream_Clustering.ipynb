```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the dataset (Assuming `click` is your DataFrame)

clickstream_train = pd.read_csv('C:/Users/sanji/Desktop/ClickStream/clickstream_train_prediction.csv')
X = clickstream_train[['month', 'day', 'session_id', 'page1_main_category', 'order', 'price','price_2']]

# Standardize the features (K-Means is sensitive to scale)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Elbow Method to find optimal K
inertia = []
K_range = range(1, 4)  # Try K from 1 to 10

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()

# Choose optimal K (for example, if the elbow is at K=3)
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
clickstream_train['Cluster'] = kmeans.fit_predict(X_scaled)

# Scatter Plot (Using first two features for visualization)
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clickstream_train['Cluster'], cmap='viridis', alpha=0.7)
plt.xlabel('Standardized Month')
plt.ylabel('Standardized Day')
plt.title('K-Means Clustering Results')
plt.colorbar(label='Cluster')
plt.show()

# Display few rows with cluster labels
print(clickstream_train[['month', 'day', 'session_id', 'page1_main_category', 'order', 'price', 'Cluster', 'price_2']].head(10))
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt

# Extract the 'order' column
order_data = clickstream_train[['order', 'price_2']]

# Standardize the 'order' column
scaler = StandardScaler()
order_scaled = scaler.fit_transform(order_data)

# Elbow Method to find optimal K
inertia = []
K_range = range(1, 3)  # Try K from 1 to 10

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(order_scaled)
    inertia.append(kmeans.inertia_)

# Plot Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K (Order Column)')
plt.show()

# Choose optimal K (for example, if the elbow is at K=3)
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
clickstream_train['Order_Cluster'] = kmeans.fit_predict(order_scaled)

# Display few rows with cluster labels
print(clickstream_train[['order','price_2','Order_Cluster']].head(10))
# Scatter Plot (Using 'order' and 'purchase' for visualization)
plt.figure(figsize=(8, 6))
plt.scatter(order_scaled[:, 0], order_scaled[:, 1], c=clickstream_train['Order_Cluster'], cmap='viridis', alpha=0.7)
plt.xlabel('Standardized Order')
plt.ylabel('Standardized Purchase')
plt.title('K-Means Clustering Results (Order Column)')
plt.colorbar(label='Cluster')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt




X = clickstream_train[['month', 'day', 'page1_main_category', 'order', 'price','price_2']]

# Initialize KMeans with k=4
kmeans = KMeans(n_clusters=4, random_state=42)  # Set random_state for reproducibility

# Fit the model to the data
kmeans.fit(X)

# Get cluster labels
# Get cluster labels
labels = kmeans.labels_

# Calculate clustering metrics
silhouette_avg = silhouette_score(X, labels)
davies_bouldin = davies_bouldin_score(X, labels)
wcss = kmeans.inertia_

print(f"Silhouette Score: {silhouette_avg}")
print(f"Davies-Bouldin Index: {davies_bouldin}")
print(f"Within-Cluster Sum of Squares (WCSS): {wcss}")

# Add cluster labels to the DataFrame
clickstream_train['Cluster'] = labels

# Visualize the clusters
scatter = plt.scatter(clickstream_train['month'], clickstream_train['day'], c=clickstream_train['Cluster'], cmap='viridis', alpha=0.7)
plt.xlabel('months')
plt.ylabel('days')
plt.title('KMeans Clustering (k=4)')
plt.colorbar(scatter, label='Cluster')
plt.show()
```
